```python
import pandas as pd
import numpy as np
import matplotlib as plt
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure, show
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource, ColorBar
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
fig = figure(title = 'sine wave')
x = np.linspace(0, 10, 1000)  
y = np.sin(x)  

p = figure(title="Sine Wave", x_axis_label="x", y_axis_label="sin(x)")
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
x = np.random.randn(100)
y = np.random.randn(100)
sizes = np.random.normal(loc = 20, scale = 10, size = 100)
colors = np.random.rand(100) * 100
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))
fig = figure(title = "Scatter plot")
fig.scatter(x='x', y='y', size='sizes',color=linear_cmap('colors', 'Viridis256', 0, 100), source=source, fill_alpha=0.6)
show(fig)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", toolbar_location=None, tools="")
p.vbar(x='fruits', top='counts', width=0.9, source=source)
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram of Normally Distributed Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency',
           tools='')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color='navy', line_color='white')
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

data_flattened = data_heatmap.flatten()
x_flattened = xx.flatten()
y_flattened = yy.flatten()

source = ColumnDataSource(data=dict(
    value=data_flattened,
    x_left=x_flattened - 0.05,
    x_right=x_flattened + 0.05,
    y_bottom=y_flattened - 0.05,
    y_top=y_flattened + 0.05,
))


p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1), tools="", width=400, height=400)


mapper = linear_cmap(field_name='value', palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

p.quad(source=source, left='x_left', right='x_right', bottom='y_bottom', top='y_top',
        line_color='white', fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], location=(0,0))
p.add_layout(color_bar, 'right')
show(p)
```
