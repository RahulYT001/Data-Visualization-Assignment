1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
dataset.

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
a = np.random.randn(10)
b = np.random.randn(10)
dataset = pd.DataFrame({'X': a, 'Y': b})
```

```python
sns.relplot(x = a, y = b, data = dataset, kind = 'scatter')
plt.xlabel('X variable')
plt.ylabel('Y variable')
plt.title('Seaborn Scatter Plot of Synthetic Data')
```

2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
x = np.random.randn(1000)

sns.histplot(x, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Random Numerical Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

3. Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.

```python
cat = ['Category A', 'Category B', 'Category C', 'Category D']
val = [20,25,30,44]
sns.barplot(x = cat, y = val)
plt.grid("True")
plt.title("Comparing values of different categories")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical
variable across different categories.

```python
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, 100),
    'Value': np.random.normal(loc=50, scale=10, size=100)
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.grid()
sns.boxplot(x='Category', y='Value', data=df)
plt.show()
```

5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python
feature_1 = np.random.normal(loc = 50, scale = 10, size = 100)
feature_2 = np.random.randn(100) + feature_1
feature_3 = 0.5 * feature_1 + feature_2

data = {"feature_1":feature_1, "feature_2":feature_2, "feature_3":feature_3}
full_data = pd.DataFrame(data)
corr_matrix = full_data.corr()


heatmap = sns.heatmap(corr_matrix)
plt.title("Correlation Matrix Heatmap")

heatmap
```
